```python
#%%
import numpy as np
import matplotlib.pyplot as plt  

nn = np.arange(0, 8) #Creo un vector de muestras, que vaya del 0 al 8. arange  me da los naturales entre 0 y 7. 

xx = 4 +  3 * np.sin((np.pi)/2 * nn ) #funcion senoidal original que da el enunciado, es la base.  

# plt.stem(nn,xx) #Acá graficamos para ver si me dio algo o no, pero no lo pide la consigna. Se usa Stem para ver los puntos del grafico y no una continua 
# plt.show()


# CÁLCULO DE LA FFT, EL MÓDULO Y LA FASE. 

FFT = np.fft.fft(xx)     
magnitud = np.abs(FFT) 
fase = np.angle(FFT)
```

Ahora realizo  el gráfico del módulo y la fase de la FFT.

```python
plt.figure()
plt.stem(nn, magnitud, linefmt='deeppink') #Uso Stem para ver los puntos del grafico y no una continua, y uso ese lien porque lo quiero en rosa, pero no acepta el comando color 
plt.title("Módulo de la FFT ")
plt.grid(True)
plt.show()
plt.figure()
plt.stem(nn, fase, linefmt='deeppink')
plt.title("Fase de la FFT ")
plt.grid(True)
plt.show()
```

Además, grafico también el Espectro de potencias.

$$
\frac{1}{N^2} \, |X[k]|^2
$$

```python
espectpot = (1/(8)**2) * (magnitud)**2 #Donde 8 es la cantidad de muestras
plt.figure()
plt.stem(espectpot, linefmt='deeppink')
plt.title("Espectro de potencia ")
plt.grid(True)
plt.show() 
```

Ahora probamos con 3/2 * Pi

Ahora repetimos con pi/3

La componente constante, DC = 4 , genera un delta (pico de energía) en K=0, mientras que la parte senoidal, genera dos picos, uno en K= k0 y otro en K = N - k0.

Donde: $$
k_0 = \frac{\omega_0 N}{2\pi}
$$

Para el primer caso donde:
$$
\omega_0 = \frac{\pi}{2}
$$

$$
k_0 = \frac{\omega_0 N}{2\pi} = \frac{(\pi/2)\cdot 8}{2\pi} = 2
$$

$$
\text{La energía se concentra en: } k = 2 \quad \text{y} \quad k = 6 \quad (8 - 2)
$$

$$
\text{Además: } k = 0 \quad \text{(por el valor medio = 4)}
$$

En el segundo caso, donde: 

$$
\omega_0 = \frac{3\pi}{2}
$$

$$
k_0 = \frac{(3\pi/2)\cdot 8}{2\pi} = 6
$$

$$
\text{Energía en: } k = 6 \quad \text{y} \quad k = 2 \quad (8 - 6)
$$

$$
\text{Además: } k = 0 \quad \text{(componente DC)}
$$

Por último, en el tercer caso ocurre algo particular ya que: 
$$
\omega_0 = \frac{\pi}{3}
$$

$$
k_0 = \frac{(\pi/3)\cdot 8}{2\pi} = \frac{4}{3}
$$

$$
k_0 \notin \mathbb{Z}
$$

La energía no se concentra en un único punto, sino que se distribuye en varios valores de K, este fenómeno se conoce como *fuga espectral*.
